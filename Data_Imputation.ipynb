```python
#we see that the missing values in the 'First Construction Document Date' correspond with the "not issued values"
#this shows that the missing values are NMAR
#construction begins only after date of issue
#fill in missing values in 'First Construction Document Date' with "Not Issued yet"
list_of_null=[]
for index in range(len(sf_building_permits)):
    if (type(sf_building_permits['First Construction Document Date'][index]) == float):
        list_of_null.append(index)
```

```python
#It is seen that some of the values are missing even after the date of issue has been given. 
list_of_missing = []
for index in list_of_null:
    if not (sf_building_permits['Issued Date'][index]) == "Not Issued yet":
        list_of_missing.append(index)
        print(index,sf_building_permits['Issued Date'][index],sf_building_permits['First Construction Document Date'][index])
```

```python
#Seperate out the NMAR values from the MAR values
for i in list_of_null:
    if i in list_of_missing:
        list_of_null.remove(i)
```

```python
#Replace the NMAR values with "Not issued yet"
sf_building_permits['First Construction Document Date'][list_of_null]="Not Issued yet"
```

```python
#frequency of missing values in each columns. We see that the "First construction Date" column has reduced to 0.1%
percent_null = sf_building_permits.isnull().sum()
(percent_null*100)/len(sf_building_permits)
```

```python
#data imputation using cross validation

import numpy as np
import pylab as pl
```

```python
#sf_building_permits_new[['Number of Existing Stories']] = sf_building_permits_1[['Number of Existing Stories']]
df = sf_building_permits[['Number of Existing Stories','Number of Proposed Stories']]
```

```python
is_test = np.random.uniform(0, 1, len(df)) > 0.75
train = df[is_test==False]
test = df[is_test==True]
```

```python
len(train), len(test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

story_imputer = KNeighborsRegressor(n_neighbors=2000)

#split our data into 2 groups; data containing nulls and data 
# not containing nulls we'll train on the latter and make
# 'predictions' on the null data to impute monthly_income
train_w_proposed_stories = train[train['Number of Proposed Stories'].isnull()==False]
train_w_null_proposed_stories = train[train['Number of Proposed Stories'].isnull()==True]
```

```python
cols = ['Number of Existing Stories']
story_imputer=story_imputer.fit(train_w_proposed_stories[cols], train_w_proposed_stories['Number of Proposed Stories'])
```

```python
new_values = story_imputer.predict(train_w_null_proposed_stories[cols])
```

```python
new_values=np.around(new_values)
```

```python
train_w_null_proposed_stories['Number of Proposed Stories'] = new_values
```

```python
train = train_w_proposed_stories.append(train_w_null_proposed_stories)
len(train)
```
